```python
import sfml.graphics as sf
import sfml.window as sfw
import time
import sys
import random
font = sf.Font()
score = -1
class Borders:
    def __init__(self):
        self.rect = sf.RectangleShape()
        self.next = None
        self.boundingBox = sf.FloatRect()
class Snake:
    def __init__(self):
        self.x = 0
        self.y = 0
        self.next = None
        self.prev = None
        self.rectangle = sf.RectangleShape()
        self.boundingBox = sf.FloatRect()
class Movement:
    def __init__(self):
        self.a = 0
        self.b = 0
        self.next = None
def construct_borders():
    rect1 = Borders()
    rect2 = Borders()
    rect3 = Borders()
    rect4 = Borders()
    rect1.rect.size = sf.Vector2(20, 410)
    rect2.rect.size = sf.Vector2(470, 10)
    rect3.rect.size = sf.Vector2(470, 10)
    rect4.rect.size = sf.Vector2(20, 410)
    rect1.rect.fill_color = sf.Color.WHITE
    rect2.rect.fill_color = sf.Color.WHITE
    rect3.rect.fill_color = sf.Color.WHITE
    rect4.rect.fill_color = sf.Color.WHITE
    rect1.rect.position = sf.Vector2(10, 70)
    rect2.rect.position = sf.Vector2(10, 70)
    rect3.rect.position = sf.Vector2(10, 470)
    rect4.rect.position = sf.Vector2(470, 70)
    rect1.next = rect2
    rect2.next = rect3
    rect3.next = rect4
    rect4.next = None
    btemp = rect1
def insert(window, status, text):
    status[0] = False
    p = Snake()
    p.rectangle.size = sf.Vector2(10, 10)
    p.rectangle.outline_thickness = -1
    p.rectangle.outline_color = sf.Color.BLACK
    if top is None:
        p.rectangle.fill_color = sf.Color.GREEN
        p.rectangle.position = sf.Vector2(240, 240)
        top = p
        last = p
        p.next = None
        p.prev = None
        p.x = 0
        p.y = -10
        temp1 = p
        window.clear(sf.Color.BLACK)
        window.draw(p.rectangle)
        window.display()
    else:
        p.rectangle.fill_color = sf.Color(0, 140, 0)
        p.next = temp1
        p.prev = last
        p.x = temp1.x
        p.y = temp1.y
        top = p
        temp1.prev = p
        position = temp1.rectangle.position
        position.y = -temp1.y + position.y
        position.x = -temp1.x + position.x
        p.rectangle.position = position
        window.clear(sf.Color.BLACK)
        temp3 = last
        while temp3 != last:
            window.draw(temp3.rectangle)
            temp3 = temp3.prev
        t = btemp
        while t is not None:
            window.draw(t.rect)
            t = t.next
        window.draw(text)
        window.display()
        temp1 = p
def update_score(score1):
    global score
    score += 1
    str = str(score)
    score1.font = font
    score1.character_size = 40
    score1.position = sf.Vector2(200, 10)
    score1.color = sf.Color.WHITE
    score1.string = str
def insert_movement(a, b):
    if last.x != -a or last.y != -b:
        n = Movement()
        n.a = a
        n.b = b
        n.next = None
        if first is None:
            temp = n
            first = n
        else:
            temp.next = n
            temp = n
def remove_movement():
    temp2 = first
    temp3 = first.next
    if first.next is not None:
        first.next = temp3.next
        first = temp3
        del temp2
    else:
        del first
        first = None
def generate_random():
    if score == 0:
        random.seed(time.time())
    x = 31 + random.randint(0, 428)
    y = 81 + random.randint(0, 383)
    position = sf.Vector2(x, y)
    dummy = sf.CircleShape(4)
    dummy.position = position
    bb = dummy.global_bounds
    temp3 = last
    while temp3 != last:
        temp3.boundingBox = temp3.rectangle.global_bounds
        if temp3.boundingBox.intersects(bb):
            position = generate_random()
        temp3 = temp3.prev
    return position
food = sf.CircleShape(4)
def generate_food():
    food.position = generate_random()
    food.fill_color = sf.Color.YELLOW
def draw_all(window, score2, score3):
    t = btemp
    while t is not None:
        window.draw(t.rect)
        t = t.next
    window.draw(food)
    window.draw(score3)
    window.draw(score2)
    temp5 = last
    while temp5 != last:
        temp5.rectangle.move(temp5.x, temp5.y)
        window.draw(temp5.rectangle)
        temp5 = temp5.prev
def collision_check(status):
    boundingBox1 = last.rectangle.global_bounds
    boundingBox2 = food.global_bounds
    if boundingBox2.intersects(boundingBox1):
        generate_food()
        status[0] = True
    temp5 = last.prev
    while temp5 != last:
        temp5.boundingBox = temp5.rectangle.global_bounds
        if temp5.boundingBox.intersects(boundingBox1):
            return True
        temp5 = temp5.prev
    t = btemp
    while t is not None:
        t.boundingBox = t.rect.global_bounds
        if t.boundingBox.intersects(boundingBox1):
            return True
        t = t.next
def moving(window):
    construct_borders()
    text = sf.Text()
    text.font = font
    text.string = "Game Over"
    text.character_size = 40
    text.position = sf.Vector2(275, 10)
    text.color = sf.Color(255, 0, 51)
    mytext = sf.Text()
    mytext.font = font
    mytext.string = "Score: "
    mytext.character_size = 40
    mytext.position = sf.Vector2(20, 10)
    score2 = update_score()
    collision = False
    status = False
    insert(window, status, mytext)
    insert(window, status, mytext)
    insert(window, status, mytext)
    insert(window, status, mytext)
    generate_food()
    while window.is_open:
        temp4 = top
        previous = temp4.next
        if first is not None:
            last.x = first.a
            last.y = first.b
            remove_movement()
        window.clear(sf.Color.BLACK)
        draw_all(window, score2, mytext)
        window.display()
        collision = collision_check(status)
        if status:
            insert(window, status, mytext)
            update_score(score2)
        if collision:
            window.draw(text)
            window.display()
            break
        time.sleep(0.1)
        temp4 = top
        while temp4 != last:
            temp4.x = previous.x
            temp4.y = previous.y
            temp4 = temp4.next
            previous = previous.next
def title_screen(window):
    font1 = sf.Font()
    if not font.load_from_file("brokenstick.ttf"):
        print("blahblah error")
        sys.exit()
    if not font1.load_from_file("BRLNSR.ttf"):
        print("blahblah error")
        sys.exit()
    logo = sf.Texture()
    if not logo.load_from_file("k11766153.png"):
        print("blahblah error")
        sys.exit()
    sprite = sf.Sprite()
    sprite.texture = logo
    sprite.position = sf.Vector2(165, 30)
    sprite.color = sf.Color(0, 240, 0)
    snake1 = sf.Text()
    snake1.font = font
    snake1.string = "Snake"
    snake1.character_size = 50
    snake1.color = sf.Color(0, 230, 0)
    snake1.position = sf.Vector2(185, 164)
    text = sf.Text()
    text.font = font1
    text.string = "Created By: "
    text.character_size = 30
    text.color = sf.Color(255, 255, 51)
    text.style = sf.Text.BOLD
    text.position = sf.Vector2(175, 230)
    text2 = sf.Text()
    text2.font = font1
    text2.string = "Ayesha Waheed-040\nDaniya Asim-041\nKomal Mariam-046\nMaryam Atiq-076"
    text2.character_size = 24
    text2.color = sf.Color(0, 255, 255)
    text2.position = sf.Vector2(157, 270)
    window.clear(sf.Color.BLACK)
    window.draw(sprite)
    window.draw(snake1)
    window.draw(text)
    window.draw(text2)
    window.display()
    time.sleep(4)
def main():
    window = sfw.RenderWindow(sfw.VideoMode(500, 500), "Snake Game")
    window.vertical_sync_enabled = True
    window.key_repeat_enabled = False
    title_screen(window)
    window.active = False
    thread = sfw.Thread(moving, window)
    thread.launch()
    event = sfw.Event()
    while window.is_open:
        while window.poll_event(event):
            if event.type == sfw.Event.CLOSED:
                window.close()
            elif event.type == sfw.Event.KEY_PRESSED:
                if sfw.Keyboard.is_key_pressed(sfw.Keyboard.LEFT):
                    insert_movement(-10, 0)
                if sfw.Keyboard.is_key_pressed(sfw.Keyboard.RIGHT):
                    insert_movement(10, 0)
                if sfw.Keyboard.is_key_pressed(sfw.Keyboard.UP):
                    insert_movement(0, -10)
                if sfw.Keyboard.is_key_pressed(sfw.Keyboard.DOWN):
                    insert_movement(0, 10)
if __name__ == "__main__":
    main()
```
